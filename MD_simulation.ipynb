```python
# To execute the GROMACS software, you should use the "gmx" command in the terminal:
%%bash
source /usr/local/gromacs/bin/GMXRC
mkdir /content/grojobs
cd /content/grojobs
gmx
```

```python
# The topology for the protein is being produced using the CHARM36 force field in our approach:
%%bash
tar -zxvf charmm36-jul2022.ff.tgz
source /usr/local/gromacs/bin/GMXRC
printf "1\n1\n"  | gmx pdb2gmx -f Protein.pdb -o receptor.gro 
```

```python
#ligand topology formation
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx editconf -f mole_ini.pdb -o mole.gro
```

```python
#We are creating a grid box around the protein:
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx editconf -f complex.gro -o new_box.gro -bt dodecahedron -d 1.0
```

```python
#We are adding water molecules into grid system to solvate the protein  :
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx solvate -cp new_box.gro -cs spc216.gro -p topol.top -o solv.gro
```

```python
# We are getting the solvated system ready to add ions:
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx grompp -f ions.mdp -c solv.gro -p topol.top -o ions.tpr
```

```python
#Our approach involves adding ions to the system in order to balance out the charge:
%%bash
source /usr/local/gromacs/bin/GMXRC
printf "SOL" | gmx genion -s ions.tpr -o solv_ions.gro -p topol.top -pname SOD -nname CLA -neutral 
```

```python
#We are performing an energy minimization step to evaluate the stability of the system and determine its suitability for further study:
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx grompp -f em.mdp -c solv_ions.gro -p topol.top -o em.tpr
```

```python
# energy minimization run
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx mdrun -deffnm em -v 
```

```python
#An index file will be created to store the coordinate information of all molecules present in the system:
%%bash
source /usr/local/gromacs/bin/GMXRC
printf "0 & ! a H*\nq\n" | gmx make_ndx -f mole.gro -o index.ndx
```

```python
# command for system selection
%%bash
source /usr/local/gromacs/bin/GMXRC
printf "3\n"  |gmx genrestr -f mole.gro -n index.ndx -o pose.itp -fc 1000 1000 1000
```

```python
# command for index file formation
%%bash
source /usr/local/gromacs/bin/GMXRC
printf "1 | 13\nq\n" | gmx make_ndx -f em.gro -o index.ndx 
```

```python
# Creatin parameter combination for NVT noramalization
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx grompp -f nvt.mdp -c em.gro -r em.gro -p topol.top -n index.ndx -o nvt.tpr
```

```python
# NVT run
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx mdrun -ntmpi 1 -ntomp 1 -deffnm nvt -v -pin on -nb gpu
```

```python
#Creatin parameter combination for NPT noramalization
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx grompp -f npt.mdp -c nvt.gro -r nvt.gro -t nvt.cpt -p topol.top -n index.ndx -o npt.tpr
```

```python
#NPT run
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx mdrun -ntmpi 1 -ntomp 1 -deffnm npt -v -pin on -nb gpu
```

```python
#Creatin parameter combination for Nproduction run
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx grompp -f md.mdp -c npt.gro -t npt.cpt -p topol.top -n index.ndx -o md_0_1.tpr
```

```python
#Production run
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx mdrun -ntmpi 1 -ntomp 1 -deffnm md_0_1 -v -pin on -nb gpu
```

```python
#Centering the protein in the simulation box after the simulation process" - This step involves bringing the protein molecule to the center of the simulation box after the MD simulation is complete. Centering the protein helps in better visualization and analysis of the simulation results.
%%bash
source /usr/local/gromacs/bin/GMXRC
printf "Protein\nSystem\n" | gmx trjconv -s md_0_1.tpr -f md_0_1.xtc -center -ur compact -pbc mol -o md_0_1_center.xtc
```

```python
# Bring protein to the center of box
%%bash
source /usr/local/gromacs/bin/GMXRC
printf "Protein\nSystem\n" | gmx trjconv -s md_0_1.tpr -f md_0_1_center.xtc -fit rot+trans -o md_0_1_fit.xtc
```

```python
#Creating parameter file of centerd protein
%%bash
source /usr/local/gromacs/bin/GMXRC
printf "Protein\nSystem\n" | gmx trjconv -s md_0_1.tpr -f md_0_1.xtc -o md_0_1_noPBC.xtc -pbc mol -center
```

```python
# RMSD value of Backbone of protein
%%bash
source /usr/local/gromacs/bin/GMXRC
printf "Backbone\nSystem\n" | gmx rms -s md_0_1.tpr -f md_0_1_center.xtc -o rmsd.xvg -tu ns
```

```python
#Radius of gyration
%%bash
source /usr/local/gromacs/bin/GMXRC
printf "1"  | gmx gyrate -s md_0_1.tpr -f md_0_1_center.xtc -o gyrate.xvg
```

```python
#Parameter file formation
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx grompp -f ie.mdp -c npt.gro -t npt.cpt -p topol.top -n index.ndx -o ie.tpr
```

```python
#perform this step if you wanrt to rerun the whole prodution run again or skip this
%%bash
source /usr/local/gromacs/bin/GMXRC
gmx mdrun -ntmpi 1 -ntomp 1 -deffnm ie -v -pin on -nb gpu -rerun md_0_1.xtc 
```

```python
#Energy of interaction
%%bash
source /usr/local/gromacs/bin/GMXRC
printf"21 | 22\n"  | gmx energy -f ie.edr -o interaction.xvg 
```
